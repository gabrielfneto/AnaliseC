```python
%%timeit
arq['imc_aval'] = "Baixo Peso"
arq.loc[(arq['imc'] >= 18.5) & (arq["imc"] < 25) , "imc_aval"] = "Normal"
arq.loc[(arq['imc'] >= 25) & (arq["imc"] < 30) , "imc_aval"] = "Sobrepeso"
arq.loc[(arq['imc'] >= 30) , "imc_aval"] = "Obesidade"


```
